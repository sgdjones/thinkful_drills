```python
# boilerplate
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# distributionvar = np.random.distributionname([arguments], 100)
```

```python
# Our basic binomial with one trial repeated 100 times
# (n,p,[size]) samples are drawn from a binomial distribution with specified parameters,
# n trials and p probability of success where n and interger >= 0
# and p is in the interval [0,1] (n maybe input as a float, but it is truncated to an interger in use)

# When estimating the standard error of a proporation in a popluation by using a random sample, 
# the normal distribution works well unless the product p*n <=5, 
#where p=populcation proporation estimate, and
# n=number of samples, in which case the binomial distribution is used instead.
### For example, a sample of 15 people show 4 who are left-handed, and 11 who are right-handed. 
### Then p=4/15=27%  0.27*15=4, so the binomial distribution should be used in this case.

binomial = np.random.binomial(1, .5, 100)

#histogram
plt.hist(binomial)


#calculating mean and std
plt.axvline(binomial.mean(), color = 'y', linestyle="solid", linewidth=2)
plt.axvline(binomial.mean() + binomial.std(), color = "y", linestyle = "dotted", linewidth=2)
plt.axvline(binomial.mean() - binomial.std(), color = "y", linestyle = "dotted", linewidth=2)
plt.title("Binomial Distribution")

plt.show()
```

```python
# bernoulli - essentially alot of binomial trials. Like 1000 coinflips, repeated 100 times.

# A Bernoulli random variable has two possible outcomes: 0 or 1.
# A binomial distribution is the sum of independent and identically distributed Bernoilli random variables.

### All Bernoulli distributions are binomial distributions, 
### but most binomial distributions are not Bernoulli distributions.

bernoulli = np.random.binomial(1000, .5, 100)

plt.hist(bernoulli)

plt.axvline(bernoulli.mean(), color = 'y')
plt.axvline(bernoulli.mean() + bernoulli.std(), color='y', linestyle='dotted')
plt.axvline(bernoulli.mean() - bernoulli.std(), color='y', linestyle='dotted')
plt.title("Bernoulli Distribution")

plt.show()
```

```python
#poisson = (lam, size)
# lam should be >=0.  A sequence of expectation intervals mush be broadcastable over the requested size
# size s/b an int or tuple of ints and is optional.  If size is None, a single value is returned if lam is a scaler
# The Poisson distribution is the limit of the binomial distribution for large N
# It shows the likely number of times that an event will occur within a specified period of time.
# It is used for independent events which occur at a constant rate within a given interval of time.

poisson = np.random.poisson(5,100)

plt.hist(poisson)

plt.axvline(poisson.mean(), color='y')
plt.axvline(poisson.mean() + poisson.std(), color='y', linestyle="dotted")
plt.axvline(poisson.mean() - poisson.std(), color='y', linestyle="dotted")

plt.show()
```

```python
# exponential = (scale=1.0, size=None)
# The exponential distribution is a continous analogue of the geometric distribution.
# It describes many common situations, such as the size of raindrops measures over many storms
# or the time between page requests to Wikipedia

exponential = np.random.exponential(30, 100)

plt.hist(exponential)

plt.axvline(exponential.mean(), color='y')
plt.axvline(exponential.mean() + exponential.std(), color='y', linestyle='dotted')
plt.axvline(exponential.mean() - exponential.std(), color='y', linestyle='dotted')
plt.title("Exponential Distribution")

plt.show()
```

```python
# trying a geometric distribution
#Described as follows: Bernoulli trials are experiments with one of two out comes: success or failure.
#The geometric distribution models the number of trials that much be run in order to achieve sucess.
#Therefore, it is supported on the positive integers (ie k= 1,2,....)

geometric = np.random.geometric(.33,100)

plt.axvline(geometric.mean(), color="y")
plt.axvline(geometric.mean() + geometric.std(), color="y", linestyle="dotted")
plt.axvline(geometric.mean() - geometric.std(), color="y", linestyle="dotted")
plt.title("Geometric Distribution")

plt.hist(geometric)
plt.show()
```

```python
#Gumbel Distribution (loc, scale, size)
# gumble(loc=0.0, scale=1.0, size=None)
# used in modeling extreme value problems.  Originally used for modeling the occurence of flood events. 
# It is also used for modeling maximum windspeed and rainfall rates.
### It is a "fat-tailed" distribution - the probability of an event in the tail distribution is larger than if one used a Guassian.
# Hence, the suprisingly frequent occurance of 100-year floods.  
# Floods were initially modeled as a Guassian process, which underestimates the frequency of extreme events.

gumbel = np.random.gumbel(0,1,100)

plt.hist(gumbel)

plt.axvline(gumbel.mean(), color="y")
plt.axvline(gumbel.mean() + gumbel.std(), color="y", linestyle="dotted")
plt.axvline(gumbel.mean() - gumbel.std(), color="y", linestyle="dotted")
plt.title("Gumbel Distribution")

plt.show()
```

```python
# Normal Distribution aka Gaussian Distribution
#Often called the bell curve because of its characteristic shape.  The normal distributions occur often in nature.  For example, it describes the commonly occuring distribution
#of samples influenced by a large number of tiny, random disturbances, which with its own unique distribution.

normal_dist_1 = np.random.normal(5, .5, 100)
normal_dist_2 = np.random.normal(10,1,100)
normal_dist_3 = normal_dist_1 + normal_dist_2

plt.axvline(normal_dist_1.mean(), color="y")
plt.axvline(normal_dist_1.mean() + normal_dist_1.std(), color="y", linestyle="dotted")
plt.axvline(normal_dist_1.mean() - normal_dist_1.std(), color="y", linestyle="dotted")

plt.axvline(normal_dist_2.mean(), color="y")
plt.axvline(normal_dist_2.mean() + normal_dist_2.std(), color="y", linestyle="dotted")
plt.axvline(normal_dist_2.mean() - normal_dist_2.std(), color="y", linestyle="dotted")

plt.axvline(normal_dist_3.mean(), color="y")
plt.axvline(normal_dist_3.mean() + normal_dist_3.std(), color="y", linestyle="dotted")
plt.axvline(normal_dist_3.mean() - normal_dist_3.std(), color="y", linestyle="dotted")
            
plt.hist(normal_dist_1, color="b", alpha=.5)
plt.hist(normal_dist_2, color="r", alpha=.5 )
plt.hist(normal_dist_3, color="g", alpha=.5)

plt.show()
```
