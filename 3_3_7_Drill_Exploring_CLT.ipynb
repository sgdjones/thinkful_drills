```python
# The boilerplate
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10, 0.5, 10000)

plt.hist(pop1, alpha=.5, label="Population 1")
plt.hist(pop2, alpha=.5, label="Population 2")
plt.legend(loc="upper right")
plt.title("Population Distribution")

plt.show()
```

```python
# Sample size of 100 for both populations
sample1 = np.random.choice(pop1, 100, replace = True)
sample2= np.random.choice(pop2, 100, replace = True)

plt.hist(sample1, alpha=0.5, label="sample 1")
plt.hist(sample2, alpha=0.5, label="sample 2")
plt.legend(loc='upper right')
plt.title("Sample Size 100")

plt.show()
```

```python
# the means and standard deviations of both samples

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())
```

```python
# Task 1
# Sample size of 1000 for both populations

sample1b = np.random.choice(pop1, 1000, replace=True)
sample2b = np.random.choice(pop2, 1000, replace=True)

plt.hist(sample1b, alpha=0.5, label="Sample 1B")
plt.hist(sample2b, alpha=0.5, label="Sample 2B")
plt.legend(loc="upper right")
plt.title("Sample Size 1000")

plt.show()
```

```python
# Task 2
# Updated pop1 probability values.  
#Expect the distribution to be more similar and overlap (reducing the variance between the two samples)

pop1_p3 = np.random.binomial(10, 0.3, 10000)
pop1_p4 = np.random.binomial(10, .4, 10000)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.hist(pop1, alpha=.5, label="Population 1p2")
plt.hist(pop2, alpha=.5, label="Population 2")
plt.legend(loc="upper right")
plt.title("Probability p.2 Population Distribution (original)")

plt.subplot(1,3,2)
plt.hist(pop1_p3, alpha=.5, label="Updated Population 1p3")
plt.hist(pop2, alpha=.5, label="Updated Population 2")
plt.legend(loc="upper right")
plt.title("Probability .3 Population Distribution")

plt.subplot(1,3,3)
plt.hist(pop1_p4, alpha=0.5, label="Population 1p4")
plt.hist(pop2, alpha=.5, label="Updated Population 2")
plt.legend(loc="upper right")
plt.title("Probability .4 Population Distribution")

plt.show()
```

```python
pop1_normal = np.random.normal(10, 0.2, 10000)
pop2_normal = np.random.normal(10, 0.5, 10000)

plt.hist(pop1_normal, alpha=.5, color = "c", label="Population 1")
plt.hist(pop2_normal, alpha=.5, color = "m", label="Population 2")
plt.legend(loc="upper right")
plt.title("Normal Population Distribution")

plt.show()
```
