```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline


df = pd.read_csv("WELLCOME_APCspend2013_forThinkful.csv", encoding = 'latin_1')
df.head(100)

#FYI - PMCID articles are publicly compliant and shared PMID may not be
```

```python
#exploring the dataset
df.describe()
```

```python
# Rename the columns
df.columns = ['reference', 'publisher', 'journal', 'article', 'cost']
```

```python
#drop any NA values
df = df.dropna()
```

```python
df.describe()
```

```python
#reference numbers for articles should be unique, so we are removing duplicates
df.drop_duplicates(['reference'])
df.drop_duplicates(['article'])

#remove any monetary symbols and convert to a float
df['cost'] = df['cost'].str.replace('$', '')
df['cost'] = df['cost'].str.replace('£', '')

#remove whitspace
df['reference'] = df['reference'].str.strip()
df['publisher'] = df['publisher'].str.strip()
df['journal'] = df['journal'].str.strip()
df['article'] = df['article'].str.strip()
df['cost'] = df['cost'].str.strip()
```
